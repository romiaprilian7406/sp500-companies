# Import Libraries

```python
import pandas as pd
import requests
import io
```

# Global Configurations

```python
# Konfigurasi URL dan User-Agent
WIKI_URL = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
HEADERS = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}
```

# Define Functions

```python
# Mengambil data mentah tabel S&P 500 dari Wikipedia
def fetch_sp500(url):
    try:
        # Request ke URL
        response = requests.get(url, headers=HEADERS)
        response.raise_for_status()

        # Parsing HTML
        tables = pd.read_html(io.StringIO(response.text))

        # Tabel utama S&P 500 ada di indeks 0
        return tables[0]

    except requests.exceptions.RequestException as e:
        print(f"Error saat mengambil data: {e}")
        raise
```

# Execution

```python
# Fetch Data
sp500_df = fetch_sp500(WIKI_URL)
print(f"Total ticker: {len(sp500_df)}")

print("\nPreview Data")
display(sp500_df.head())
```

# Export File

```python
file_name = "sp500_companies.csv"
sp500_df.to_csv(file_name, index=False)
print(f"File berhasil disimpan sebagai: {file_name}")
```
